```python
import pandas as pd
import numpy as np
```

```python
import urllib.request
import os
os.getcwd()
```

```python
file_name = 'student_resource_3/dataset/sample_test.csv'
file_path = 'student_resource_3/dataset/sample_images'
```

```python
def url_to_jpg(i, url, file_path):
        file_name = 'img-{}.jpg'.format(i)
        full_path = os.path.join(file_path, file_name)
        print(f"Downloading image {i} from {url} to {full_path}")
        
        urllib.request.urlretrieve(url, full_path)
        print(f"Image {i} successfully saved!")
```

```python
def download_images_from_csv(csv_file, file_path):
    # Read CSV file
    df = pd.read_csv(file_name)
    
    # Download images
    for i, url in enumerate(df['image_link']):
        url_to_jpg(i, url, file_path)
```

```python
download_images_from_csv(file_name, file_path)
```
